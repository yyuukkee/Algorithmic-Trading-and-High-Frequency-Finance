```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
file_path = '/content/drive/My Drive/deaths-by-day.csv'
df = pd.read_csv(file_path)

df.head()
```

```python
y = df['DEATH_COUNT']

X_lin = np.arange(len(y)).reshape(-1, 1)
y_lin = y.values

model_lin = LinearRegression()
model_lin.fit(X_lin, y_lin)
y_pred_lin = model_lin.predict(X_lin)

resid_lin = y_lin - y_pred_lin

plt.figure(figsize=(10, 4))
plt.plot(resid_lin, label='Linear Regression Residuals')
plt.title('Residuals of Linear Regression')
plt.legend()
plt.grid(True)
plt.show()

X_ar = y[:-1].values
y_ar = y[1:].values

X_ar = sm.add_constant(X_ar)
model_ar = sm.OLS(y_ar, X_ar).fit()
y_pred_ar = model_ar.predict(X_ar)

resid_ar = y_ar - y_pred_ar

plt.figure(figsize=(10, 4))
plt.plot(resid_ar, label='AR(1) Residuals', color='orange')
plt.title('Residuals of AR(1) Model')
plt.legend()
plt.grid(True)
plt.show()

print("Linear Regression R²:", model_lin.score(X_lin, y_lin))
print("AR(1) Model Summary:")
print(model_ar.summary())
```

The AR(1) model is preferred over linear regression because it provides a much better fit to the data. It achieves a significantly higher R-squared value (0.987 vs. 0.121), indicating that it explains more of the variance in daily COVID deaths. Additionally, the residuals from the AR(1) model are smaller and more stable, suggesting a better capture of the time series structure. In contrast, the linear regression residuals show clear patterns, indicating model misspecification.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def moving_average(x, window=4):
    return np.convolve(x, np.ones(window)/window, mode='valid')
```

```python
np.random.seed(0)
n = 100
w = np.random.normal(0, 1, n)

x_a = np.zeros(n)
x_a[0:2] = w[0:2]
for t in range(2, n):
    x_a[t] = -0.9 * x_a[t-2] + w[t]

v_a = moving_average(x_a)
```

```python
t_vals = np.arange(n)
x_b = np.cos(2 * np.pi * t_vals / 4)
v_b = moving_average(x_b)
```

```python
x_c = x_b + w
v_c = moving_average(x_c)
```

```python
fig, axs = plt.subplots(3, 1, figsize=(12, 10), sharex=True)

# (a)
axs[0].plot(x_a, label='x_t (AR)')
axs[0].plot(np.arange(3, n), v_a, 'r--', label='v_t (Moving Avg)')
axs[0].set_title('(a) AR Process: $x_t = -0.9 x_{t-2} + w_t$')
axs[0].legend()

# (b)
axs[1].plot(x_b, label='x_t = cos(2πt/4)')
axs[1].plot(np.arange(3, n), v_b, 'r--', label='v_t (Moving Avg)')
axs[1].set_title('(b) Pure Cosine Signal')
axs[1].legend()

# (c)
axs[2].plot(x_c, label='x_t = cos(2πt/4) + w_t')
axs[2].plot(np.arange(3, n), v_c, 'r--', label='v_t (Moving Avg)')
axs[2].set_title('(c) Noisy Cosine Signal')
axs[2].legend()

plt.xlabel('Time')
plt.tight_layout()
plt.show()
```

(a)-(c)

In case (a), the autoregressive process
  exhibits oscillatory behavior with alternating signs due to the negative coefficient, and a moderate level of randomness from the white noise. Applying the moving average filter smooths the short-term fluctuations, reducing noise while still preserving the general oscillatory pattern. In case (b), the pure cosine signal is perfectly periodic. However, the moving average filter flattens this pattern, producing a near-zero line because the window spans a full cycle, effectively canceling out the periodic components. In case (c), where noise is added to the cosine signal, the filter performs best: it successfully reduces the noise while retaining the underlying periodic structure. Overall, the moving average filter is most effective in denoising signals like (c) that contain a clear underlying structure but are corrupted by random noise.

(d)

In (a), the AR process is noisy and irregular; the moving average smooths it but doesn't reveal a clear pattern. In (b), the cosine wave is perfectly periodic, but the filter flattens it due to averaging over a full cycle. In (c), the cosine wave with added noise is cleaned up effectively by the filter, revealing the underlying periodic structure. Overall, the filter works best in (c), where it reduces noise without losing the signal.
